```
def ResetBoard():
  board = [[0 for x in range(9)] for y in range(9)]
  return board

def DrawBoard(board):
  i = 0
  for row in board:
    print(*row[0:3], sep=' ', end='')
    print('  |  ', end='')
    print(*row[3:6], sep=' ', end='')
    print('  |  ', end='')
    print(*row[6:9], sep=' ')
    if i == 2 or i == 5:
      print("_________________________\n")
    i+=1
  print('\n')

def IsInRow(board, y, x, i):
  if i in board[x]:
    return True
  return False

def IsInCol(board, y, x, i):
  tempCol = []
  for k in range(9):
    tempCol.append(board[k][y])
  if i in tempCol:
    return True
  return False

def IsInBox(board, y, x, i):
  #given (x,y), determine which of 9 boxes its in
  #predefine each box:
  tempBox = []
  box = WhichBox(y, x)
  for item in box:
    j, k = item
    tempBox.append(board[j][k])
  if i in tempBox:
    return True
  return False

def IsMoveLegal(board, y, x, i):
  if board[x][y] is 0:
    if not IsInRow(board, y, x, i) and not IsInBox(board, y, x, i) and not IsInCol(board, y, x, i):
      return True
  return False

def WhichBox(y, x):
  box1 = [(0,0),(0,1),(0,2),(1,0),(1,1),(1,2),(2,0),(2,1),(2,2)]
  box2 = [(3,0),(3,1),(3,2),(4,0),(4,1),(4,2),(5,0),(5,1),(5,2)]
  box3 = [(6,0),(6,1),(6,2),(7,0),(7,1),(7,2),(8,0),(8,1),(8,2)]
  box4 = [(0,3),(0,4),(0,5),(1,3),(1,4),(1,5),(2,3),(2,4),(2,5)]
  box5 = [(3,3),(3,4),(3,5),(4,3),(4,4),(4,5),(5,3),(5,4),(5,5)]
  box6 = [(6,3),(6,4),(6,5),(7,3),(7,4),(7,5),(8,3),(8,4),(8,5)]
  box7 = [(0,6),(0,7),(0,8),(1,6),(1,7),(1,8),(2,6),(2,7),(2,8)]
  box8 = [(3,6),(3,7),(3,8),(4,6),(4,7),(4,8),(5,6),(5,7),(5,8)]
  box9 = [(6,6),(6,7),(6,8),(7,6),(7,7),(7,8),(8,6),(8,7),(8,8)]
  if (x,y) in box1:
    return box1
  if (x,y) in box2:
    return box2
  if (x,y) in box3:
    return box3
  if (x,y) in box4:
    return box4
  if (x,y) in box5:
    return box5
  if (x,y) in box6:
    return box6
  if (x,y) in box7:
    return box7
  if (x,y) in box8:
    return box8
  if (x,y) in box9:
    return box9

def CreateMoveList(board):
  moveList = [[0 for x in range(9)] for y in range(9)]
  tempList = []
  for j in range (9):
    for k in range(9):
      for i in range(1,10):
        if IsMoveLegal(board, k, j, i):
          tempList.append(i)
      moveList[j][k] = copy.copy(tempList)
      tempList = []
  return moveList

def FindOptimalMove(moveList):
  for j in range (9):
    for k in range(9):
      if len(moveList[k][j]) is 1:
        i = moveList[k][j]
        return (j, k, i[0])
  return (0,0,0)

def CheckSolved(board):
  checkList = []
  for x in range(9):
    for y in range(9):
      checkList.append(board[x][y])
  if sum(checkList) == 405:
    return True
  return False

def MakeMove(board, y, x, i):
  if i is not 0:
    board[x][y] = i

def SolveSudoku():
  # Uncomment '#' the Puzzle you would like to see solved.
  # Don't run Hard or Evil boards, they break the program with an infinite loop currently.
  
  board = SetEasyBoard()
  # board = SetMediumBoard()
  # board = SetHardBoard()
  # board = SetEvilBoard()
  print("Original Puzzle State:")
  DrawBoard(board)
  moves = 1
  while not CheckSolved(board):
    y, x, i = FindOptimalMove(CreateMoveList(board))
    MakeMove(board, y, x, i)
    print('Moves:', moves)
    DrawBoard(board)
    moves+=1
  print("Done! =D")
```
